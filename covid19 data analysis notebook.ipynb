### Task 2.3: Aggregating the rows by the country

```python
corona_aggregated_dataset = corona_dataset_csv.groupby("Country/Region").sum()
```

### task 3.1: caculating the first derivative of the curve

```python
corona_aggregated_dataset.loc["China"].diff().plot()
```

### task 3.2: find maxmimum infection rate for China

```python
corona_aggregated_dataset.loc["China"].diff().max()
```

```python
corona_aggregated_dataset.loc["Spain"].diff().max()
```

```python
corona_aggregated_dataset.loc["Italy"].diff().max()
```

### Task 3.3: find maximum infection rate for all of the countries. 

```python
countries = list(corona_aggregated_dataset.index)
max_infection_rates = []
for c in countries:
    max_infection_rates.append(corona_aggregated_dataset.loc[c].diff().max())
corona_aggregated_dataset["max_infection_rates"] = max_infection_rates
```

```python
corona_aggregated_dataset
```

### Task 3.4: create a new dataframe with only needed column 

```python
corona_data = pd.DataFrame(corona_aggregated_dataset["max_infection_rates"])
```

```python
corona_data.head()
```

### Task 5.1: Plotting GDP vs maximum Infection rate

```python
# Calculate the logarithm of max_infection_rates
data['log_max_infection_rates'] = np.log(data['max_infection_rates'])

# Plot using Seaborn scatterplot with some customizations
sns.scatterplot(data=data, x="GDP per capita", y="log_max_infection_rates", color='blue', alpha=0.7)
plt.title('Scatter Plot of GDP per capita vs Log of Max Infection Rates')
plt.xlabel('GDP per capita')
plt.ylabel('Log of Max Infection Rates')

plt.show()
```

```python
sns.regplot(data=data, x="GDP per capita", y=np.log(data["max_infection_rates"]))
```
